```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
# 2차원 데이터를 생성
X, y = make_blobs(random_state = 1)
# 군집 모델을 만듭니다 (클러스터를 2개로 했을 때)
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
print(kmeans.labels_)
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
# 2차원 데이터를 생성
X, y = make_blobs(random_state = 1)
# 군집 모델을 만듭니다 (클러스터를 5개로 했을 때)
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
print(kmeans.labels_)
```

```python
# K-means 장점
# - 이해와 구현이 쉬움, 비교적 빠름
# - 대규모 데이터셋에 적용가능
# - MiniBatchKMeans : 대용량 데이터셋 대상

# K-means 단점
# - 클러스터 모양에 제한적임 : 둥근 클러스터 모양에 잘 작동
# - 클러스터 개수를 설정 해야 함 (실제로는 알 수 없음)
# - 초기값에 따라 결과가 달라질 수 있음
```

```python
#병합군집
# - 각 데이터가 하나의 클러스터
# - 지정된 클러스터 개수가 될 때 까지 가까운 두 클러스터를 병합
from sklearn.cluster import AgglomerativeClustering
AgglomerativeClustering(n_clusters=2, affinity='euclidean', memory=None, connectivity=None, compute_full_tree='auto' ,linkage='ward')
```

```python
from scipy.cluster.hierarchy import dendrogram, ward
linkage_array = ward(X)
# 배열 X를 입력받아 연결 배열 반환
dendrogram(linkage_array)
# 세로 축 : 클러스터 사이의 거리
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
dbscan = DBSCAN() # eps = 0.5, 두 개로 잘 분리
clusters = dbscan.fit_predict(X_scaled)
# Eps를 작게 / 더 많은 클러스터 생성
dbscan = DBSCAN(eps=0.2)
# Eps를 크게 / 하나의 클러스터 생성
dbscan = DBSCAN(eps=0.7)
```

```python
# DBSCAN 장점
# - 원하는 수만큼 클러스터를 검색할 수 있다.
# - 다양한 크기 및 모양의 클러스터를 찾을 수 있다.
# - 특이치를 탐지하거나 무시할 수 있다.
# DBSCAN 단점
# - 이웃 수(사이즈)에 민감하다.
```
