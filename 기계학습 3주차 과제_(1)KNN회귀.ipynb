```python
from preamble import *
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import mglearn
%matplotlib inline
#필요한 라이브러리들을 임포트한다.

from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
#한글 출력시 오류가 날 때 추가하는 코드

X, y = mglearn.datasets.make_wave(n_samples=40)
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("특성")
plt.ylabel("타깃")
# 회귀분석을 하기 앞에서 배웠던 make_wave를 사용
```

```python
from sklearn.neighbors import KNeighborsRegressor
# KNeighborRegressor를 임포트한다.

from sklearn.model_selection import train_test_split
# train_test_split를 임포트한다.

X, y = mglearn.datasets.make_wave(n_samples=40)
#데이터 생성

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
#wave 데이터셋을 훈련 세트와 테스트 세트로 나눈다.
#(test_size의 기본값은 0.25이다.)

reg = KNeighborsRegressor(n_neighbors=3)
#이웃의 수를 3으로 하여 모델의 객체 생성

reg.fit(X_train, y_train)
#훈련 데이터와 타깃을 사용해 모델을 학습시킨다.

print("테스트 세트 예측:\n{}".format(reg.predict(X_test)))
# 테스트 세트 예측 값을 출력시켜본다.
# 이 값들은 위의 그래프에서 타깃으로 표현되는 값들을 말한다.

print("테스트 세트 R^2: {:.2f}".format(reg.score(X_test, y_test)))
# 결정 계수 출력 (추후 다뤄질 예정)
```

```python
#위에서 했던 과정을 똑같이 하는데 대신 test_size를 0.4로 한 경우
X, y = mglearn.datasets.make_wave(n_samples=40)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.4)

reg = KNeighborsRegressor(n_neighbors=3)

reg.fit(X_train, y_train)

print("(test_size=0.4) 테스트 세트 예측:\n{}".format(reg.predict(X_test)))

print("(test_size=0.4) 테스트 세트 R^2: {:.2f}\n".format(reg.score(X_test, y_test)))

#위에서 했던 과정을 똑같이 하는데 대신 test_size를 0.7로 한 경우
X, y = mglearn.datasets.make_wave(n_samples=40)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.7)

reg = KNeighborsRegressor(n_neighbors=3)

reg.fit(X_train, y_train)

print("(test_size=0.7) 테스트 세트 예측:\n{}".format(reg.predict(X_test)))

print("(test_size=0.7) 테스트 세트 R^2: {:.2f}\n".format(reg.score(X_test, y_test)))

#위에서 했던 과정을 똑같이 하는데 대신 test_size를 0.1로 한 경우
X, y = mglearn.datasets.make_wave(n_samples=40)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.1)

reg = KNeighborsRegressor(n_neighbors=3)

reg.fit(X_train, y_train)

print("(test_size=0.1) 테스트 세트 예측:\n{}".format(reg.predict(X_test)))

print("(test_size=0.1) 테스트 세트 R^2: {:.2f}".format(reg.score(X_test, y_test)))

# 차례대로 비교해보니 test_size 값이 클 수록 테스트 세트 R^2값이
# 더 작아진다는 것을 유추할 수 있었다.
```

```python
#위에서 했던 과정을 똑같이 하는데 대신 random_state를 21로 한 경우
X, y = mglearn.datasets.make_wave(n_samples=40)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=21)

reg = KNeighborsRegressor(n_neighbors=3)

reg.fit(X_train, y_train)

print("(random_state=21) 테스트 세트 예측:\n{}".format(reg.predict(X_test)))

print("(random_state=21) 테스트 세트 R^2: {:.2f}\n".format(reg.score(X_test, y_test)))

#위에서 했던 과정을 똑같이 하는데 대신 random_state를 33으로 한 경우
X, y = mglearn.datasets.make_wave(n_samples=40)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=33)

reg = KNeighborsRegressor(n_neighbors=3)

reg.fit(X_train, y_train)

print("(random_state=33) 테스트 세트 예측:\n{}".format(reg.predict(X_test)))

print("(random_state=33) 테스트 세트 R^2: {:.2f}".format(reg.score(X_test, y_test)))

#이를 통해 random_state의 변경에 따라 테스트 세트의 예측 값과 R^2값이 달라진다는 것을 알 수 있다.
```

```python
#위에서 했던 과정을 똑같이 하는데 대신 k값(n_neighbors=ㅁ에서 ㅁ안에 들어갈 값)을 줄여본 경우
X, y = mglearn.datasets.make_wave(n_samples=40)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

reg = KNeighborsRegressor(n_neighbors=1)

reg.fit(X_train, y_train)

print("(n_neighbors=1) 테스트 세트 예측:\n{}".format(reg.predict(X_test)))

print("테스트 세트 R^2: {:.2f}\n".format(reg.score(X_test, y_test)))

#위에서 했던 과정을 똑같이 하는데 대신 k값을 더 키워본 경우
X, y = mglearn.datasets.make_wave(n_samples=40)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

reg = KNeighborsRegressor(n_neighbors=5)

reg.fit(X_train, y_train)

print("(n_neighbors=5) 테스트 세트 예측:\n{}".format(reg.predict(X_test)))

print("테스트 세트 R^2: {:.2f}".format(reg.score(X_test, y_test)))

#k값의 변경도 테스트 세트의 예측 값과 R^2값의 영향을 준다는 것을 확인할 수 있었다.
```

```python
fig, axes = plt.subplots(1, 3, figsize = (15, 4))
for n_neighbors, ax in zip([1, 3, 9], axes):
    #1, 3, 9 이웃을 사용한 예측을 합니다.
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    print("X_train.shape: {}".format(X_train.shape), "X_test.shape:{}".format(X_test.shape))
    reg.fit(X_train, y_train)
    ax.plot(X_test, reg.predict(X_test), 'v', c=mglearn.cm3(2), markersize=8)
    ax.plot(X_train, y_train, '^', c=mglearn.cm3(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm3(1), markersize=8)
        
    ax.set_title("{} 이웃의 훈련 스코어: {:.2f} 테스트 스코어: {:.2f}".format(
    n_neighbors, reg.score(X_train, y_train), reg.score(X_test, y_test)))
    
    ax.set_xlabel("특성");
    ax.set_ylabel("타깃");
axes[0].legend(["모델 예측", "훈련 데이터/타깃", "테스트 데이터/타깃"], loc="best")

# 파이썬의 for문을 이용해서 이웃수를 1,3,9로 바꿔가면서 보는 그래프의 형태
# k가 증가함에 따라 훈련에 대한 정확도는 감소하고 테스트에 대해서는 더 안정된 예측을 하게 된다.
```
