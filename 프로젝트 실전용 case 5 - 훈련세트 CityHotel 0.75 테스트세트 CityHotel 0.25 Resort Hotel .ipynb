```python
import pandas as pd

Hotel_Bookings = pd.read_csv("hotel_bookings.csv")

Hotel_Bookings = Hotel_Bookings.drop(["lead_time","arrival_date_week_number","arrival_date_day_of_month"
                                     ,"country","previous_bookings_not_canceled","agent"
                                     ,"company","days_in_waiting_list","adr","reservation_status","reservation_status_date"], axis=1)
name_mapping_Month = {
    'January': 1,
    'February': 2, 
    'March': 3, 
    'April': 4, 
    'May': 5, 
    'June': 6,
    'July': 7,
    'August': 8, 
    'September': 9, 
    'October': 10, 
    'November': 11, 
    'December': 12
}
Hotel_Bookings = Hotel_Bookings.replace({'arrival_date_month':name_mapping_Month})

name_mapping_Meal = {
    'Undefined': 0,
    'BB': 1,
    'FB': 2, 
    'HB': 3, 
    'SC': 4
}
Hotel_Bookings = Hotel_Bookings.replace({'meal':name_mapping_Meal})

name_mapping_MarketSegment = {
    'Undefined': 0,
    'Aviation': 1, 
    'Complementary': 2, 
    'Corporate': 3,
    'Direct': 4,
    'Groups': 5, 
    'Offline TA/TO': 6, 
    'Online TA': 7
}
Hotel_Bookings = Hotel_Bookings.replace({'market_segment':name_mapping_MarketSegment})

name_mapping_DistributionChannel = {
    'Undefined': 0,
    'Corporate': 1, 
    'Direct': 2, 
    'GDS': 3,
    'TA/TO': 4
}
Hotel_Bookings = Hotel_Bookings.replace({'distribution_channel':name_mapping_DistributionChannel})

name_mapping_ReservedRoomType = {
    'A': 0,
    'B': 1, 
    'C': 2, 
    'D': 3,
    'E': 4,
    'F': 5, 
    'G': 6,
    'H': 7,
    'L': 8,
    'P': 9
}
Hotel_Bookings = Hotel_Bookings.replace({'reserved_room_type':name_mapping_ReservedRoomType})

name_mapping_AssignedRoomType = {
    'A': 0,
    'B': 1, 
    'C': 2, 
    'D': 3,
    'E': 4,
    'F': 5, 
    'G': 6,
    'H': 7,
    'I': 8,
    'K': 9,
    'L': 10,
    'P': 11
}
Hotel_Bookings = Hotel_Bookings.replace({'assigned_room_type':name_mapping_AssignedRoomType})

name_mapping_DepositType = {
    'No Deposit': 0,
    'Non Refund': 1, 
    'Refundable': 2
}
Hotel_Bookings = Hotel_Bookings.replace({'deposit_type':name_mapping_DepositType})

name_mapping_CustomerType = {
    'Contract': 0,
    'Group': 1, 
    'Transient': 2,
    'Transient-Party': 3
}
Hotel_Bookings = Hotel_Bookings.replace({'customer_type':name_mapping_CustomerType})

name_mapping_ReservationStatus = {
    'Canceled': 0,
    'Check-Out': 1, 
    'No-Show': 2
}
Hotel_Bookings = Hotel_Bookings.replace({'reservation_status':name_mapping_ReservationStatus})

# Hotel_Bookings.to_csv('Rata.csv',index=False,sep='\t')
```

```python
Hotel_Bookings.head()
```

```python
ResortHotel_Bookings = Hotel_Bookings[(Hotel_Bookings["hotel"] == "Resort Hotel")]

CityHotel_Bookings = Hotel_Bookings[(Hotel_Bookings["hotel"] == "City Hotel")]
CityHotel_Bookings = CityHotel_Bookings.reset_index()
CityHotel_Bookings = CityHotel_Bookings.drop(["index"], axis=1)


ResortHotel_Bookings.head()
```

```python
CityHotel_Bookings = Hotel_Bookings[(Hotel_Bookings["hotel"] == "City Hotel")]
CityHotel_Bookings = CityHotel_Bookings.reset_index()
CityHotel_Bookings = CityHotel_Bookings.drop(["index"], axis=1)
CityHotel_Bookings.head(5)
```

```python
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

test_size = 0.25
```

```python
print("arrival_date_year\n")

ABC = CityHotel_Bookings["arrival_date_year"].to_numpy()
ABC = ABC.reshape(-1,1)
Is_Canceled = np.ravel(CityHotel_Bookings["is_canceled"].to_numpy(), order="C")

X_train, X_test, y_train, y_test = train_test_split(ABC, Is_Canceled, stratify=Is_Canceled, test_size=test_size, random_state=0)

GauNB = GaussianNB()
GauNB.fit(X_train, y_train)

DTC = DecisionTreeClassifier(random_state=0)
DTC.fit(X_train, y_train)

RFC =  RandomForestClassifier(n_estimators=5, random_state=0)
RFC.fit(X_train, y_train)

GBC = GradientBoostingClassifier(random_state=0)
GBC.fit(X_train, y_train)

QQQ = ResortHotel_Bookings["arrival_date_year"].to_numpy()
QQQ = QQQ.reshape(-1,1)
Is_Canceled_City = np.ravel(ResortHotel_Bookings["is_canceled"].to_numpy(), order="C")

Fusion = np.vstack([X_test, QQQ])
Fusion_IsCanceled = np.hstack([y_test, Is_Canceled_City])

print("GaussianNB으로 계산했을 때의 정확도 : {:.4f}\n".format(GauNB.score(Fusion, Fusion_IsCanceled)))

print("DecisionTreeClassifier으로 계산했을 때의 정확도 : {:.4f}\n".format(DTC.score(Fusion, Fusion_IsCanceled)))

print("RandomForestClassifier으로 계산했을 때의 정확도 : {:.4f}\n".format(RFC.score(Fusion, Fusion_IsCanceled)))

print("GradientBoostingClassifier으로 계산했을 때의 정확도 : {:.4f}".format(GBC.score(Fusion, Fusion_IsCanceled)))
```
