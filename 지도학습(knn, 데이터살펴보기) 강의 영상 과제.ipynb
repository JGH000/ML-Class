```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()
#데이터 로드
```

```python
print("X.shape: {}".format(X.shape), "y.shape:{}".format(y.shape))
#측정치 개수 확인
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
#훈련/테스트 분할, 나누는 비율 : 75대 25
```

```python
print("X.shape: {}".format(X_train.shape), "X_test.shape:{}".format(X_test.shape))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
#모델 선택 
```

```python
clf.fit(X_train, y_train)
#훈련기 학습
```

```python
print("테스트 세트 예측: {}".format(clf.predict(X_test)))
#테스트데이터
```

```python
print("테스트 세트 예측: {:.2f}".format(clf.score(X_test, y_test)))
#성능 확인
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
#밑의 코드와 형태는 동일하지만 n_neighbors값이 달라 서로 다르게 그려진다.
#선의 개수에 주목
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10,3))
#밑의 그래프 개수를 더 추가해서 그리고 싶으면 위의 3값에서 더 큰 값으로 바꿔주면 된다.
#그리고 각 그래프의 모양은 figsize에 들어있는 값을 조절해주면 된다.
#10이 가로크기, 3이 세로크기를 결정

for n_neighbors, ax in zip([1, 3, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5,ax=ax,alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} 이웃".format(n_neighbors))
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 1")
    axes[0].legend(loc=3)
    
# n_neighbors의 값에 따라 모양이 달라지는데
# 그 값이 클 수록 경계가 부드러워진다. (포인트를 많이 참조할수록 더 일반화가 된다.)

# Bias가 낮고 Variance가 높으면 높은 복잡도를 보여주고
# 반대로 Bias가 높고 Variance가 낮으면 낮은 복잡도를 보여준다.
```

```python
fig, axes = plt.subplots(1, 5, figsize=(20,5))
#위의 그래프에서 그래프의 개수를 늘리고 그래프 크기를 변경한 것
#밑의 zip안에 배열도 변경하였다.

for n_neighbors, ax in zip([1, 3, 5, 7, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5,ax=ax,alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} 이웃".format(n_neighbors))
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 1")
    axes[0].legend(loc=3)
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print("cancer.keys(): {}".format(cancer.keys()))
```

```python
print("Shape: {}".format(cancer.data.shape))
```

```python
print("클래스별 샘플 갯수:\n{}".format({n:v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}))
```

```python
print("특성이름:\n{}".format(cancer.feature_names))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []

#1에서 10까지 n_neighbors를 적용
neighbors_settings = range(1,11)

for n_neighbors in neighbors_settings:
    #모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    
    #훈련세트 정확도 저장
    training_accuracy.append(clf.score(X_train, y_train))
    
    #일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="훈련 정확도")
plt.plot(neighbors_settings, test_accuracy, label="테스트 정확도")
plt.ylabel("정확도")
plt.xlabel("n_neighbors")
plt.legend()

#n_neighbors 증가에 따라 훈련 정확도는 점점 줄어들고 테스트정확도는 증가하다가 감소한다.
```

```python
X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify=cancer.target, random_state=42)

training_accuracy = []
test_accuracy = []

#1에서 10까지 n_neighbors를 적용
neighbors_settings = range(1,11)

for n_neighbors in neighbors_settings:
    #모델 생성
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    
    #훈련세트 정확도 저장
    training_accuracy.append(clf.score(X_train, y_train))
    
    #일반화 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="훈련 정확도")
plt.plot(neighbors_settings, test_accuracy, label="테스트 정확도")
plt.ylabel("정확도")
plt.xlabel("n_neighbors")
plt.legend()

#위의 코드에서 random_state만 42로 바꾼 코드
```
